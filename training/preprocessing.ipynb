# Feature Engineering

```python
final_df = pd.read_csv("../data/processed/final_dataset.csv")
final_df['date'] = pd.to_datetime(final_df['date'])
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])

final_df['day_of_week'] = final_df['date'].dt.dayofweek
final_df['month'] = final_df['date'].dt.month
```

```python
final_df['is_weekend'] = final_df['day_of_week'].isin([5, 6]).astype(int)
```

```python
final_df.head()
```

```python
final_df['is_hot'] = (final_df['temperature'] > 30).astype(int)
final_df['is_rainy'] = (final_df['rain'] > 0).astype(int)
```

```python
final_df['promo'].value_counts()
```

```python
final_df = final_df.sort_values(by=['store_id', 'item_id', 'date'])

final_df['lag_1'] = final_df.groupby(['store_id', 'item_id'])['sales'].shift(1)
final_df['lag_7'] = final_df.groupby(['store_id', 'item_id'])['sales'].shift(7)
```

```python
final_df['rolling_mean_7'] = final_df.groupby(['store_id', 'item_id'])['sales'].transform(lambda x: x.rolling(7).mean())
```

```python
final_df = final_df.dropna()
```

```python
final_df.head()
```

```python
# weekday and  day_of_week are same features so we can drop one of them. I am dropping weekday.
final_df = final_df.drop(columns=['weekday'])
```

```python
final_df = final_df.dropna()
```

# ML Model (forecasting Engine)

We will predict future sales

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
X = final_df.drop(columns=['sales'])
y = final_df['sales']
```

```python
final_df.columns
```

```python
final_df = final_df.sort_values('date')
```

```python
#time based split
train_size = int(len(final_df) * 0.8)

train = final_df[:train_size]
test = final_df[train_size:]
```

```python
#define x and y for train and test
X_train = train.drop(columns=['sales', 'date'])
y_train = train['sales']

X_test = test.drop(columns=['sales', 'date'])
y_test = test['sales']
```

```python
#encode categorical variables
X_train['store_id'] = X_train['store_id'].astype('category').cat.codes
X_train['item_id'] = X_train['item_id'].astype('category').cat.codes

X_test['store_id'] = X_test['store_id'].astype('category').cat.codes
X_test['item_id'] = X_test['item_id'].astype('category').cat.codes
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100),
    "XGBoost": XGBRegressor(n_estimators=100)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)

    results.append({
        "Model": name,
        "MAE": mae
    })
```

```python
import pandas as pd
pd.DataFrame(results)
```

We benchmarked multiple models including Linear Regression, Decision Trees, Random Forest, and XGBoost. While XGBoost achieved the best performance with the lowest MAE, Random Forest also performed competitively, indicating strong feature engineering and robust data representation.”

```python
# food segment
def food_segment(price):
    if price < 150:
        return "fast_moving"
    elif price < 400:
        return "regular"
    else:
        return "premium"
```

```python
#apparel segment
def apparel_segment(price):
    if price < 500:
        return "budget"
    elif price < 2000:
        return "mid_range"
    else:
        return "premium"
```

```python
#electronics segment
def electronics_segment(price):
    if price < 1000:
        return "accessories"
    elif price < 10000:
        return "mid_devices"
    else:
        return "premium_devices"
```

```python
#grocery segment
def grocery_segment(price):
    if price < 100:
        return "daily"
    elif price < 500:
        return "bulk"
    else:
        return "premium"
```

```python
def get_segment(price, category):

    if category == "food":
        return food_segment(price)

    elif category == "clothing":
        return apparel_segment(price)

    elif category == "electronics":
        return electronics_segment(price)

    elif category == "grocery":
        return grocery_segment(price)

    else:
        return "general"
```

```python
def generate_smart_action(sample, pred, category):

    price = sample['price'].values[0]
    last_sales = sample['lag_1'].values[0]
    segment = get_segment(price, category)

    diff = pred - last_sales

    if category == "clothing":
        if segment == "budget":
            return f"Increase stock of budget wear (+{int(diff)})"
        elif segment == "premium":
            return "Avoid overstocking premium items"
        else:
            return f"Focus on trending mid-range fashion (+{int(diff)})"

    elif category == "food":
        if segment == "fast_moving":
            return f"Increase fast-selling items (+{int(diff)})"
        else:
            return "Focus on quality over quantity"

    else:
        return f"Adjust stock by {int(diff)} units"
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
```

```python
import requests

def generate_ai_insight(sample, pred, category):
    
    url = "https://openrouter.ai/api/v1/chat/completions"
    
    headers = {
        "Authorization": f"Bearer {OPENROUTER_API_KEY}",
        "Content-Type": "application/json"
    }

    last_sales = sample['lag_1'].values[0]
    trend = sample['rolling_mean_7'].values[0]
    price = sample['price'].values[0]

    prompt = f"""
    You are a business advisor helping small shopkeepers.

    Business Type: {category}

    DATA:
    - Expected sales tomorrow: {pred} items
    - Sales yesterday: {last_sales} items
    - Average weekly sales: {trend} items
    - Product price: ₹{price}

    TASK:
    Explain this in simple language for a shopkeeper.

    INSTRUCTIONS:
    - Clearly explain what these numbers mean
    - Compare with yesterday and weekly average
    - Tell whether demand is increasing, decreasing, or stable
    - Give a clear action (what to do)

    STYLE:
    - Use simple language (no technical terms)
    - Speak like advising a local shop owner
    - Keep it 2–3 lines max
    """

    data = {
        "model": "openai/gpt-4o-mini",
        "messages": [
            {"role": "user", "content": prompt}
        ]
    }

    response = requests.post(url, headers=headers, json=data)
    result = response.json()

    return result['choices'][0]['message']['content']
```

```python
def full_pipeline(sample, model, category):

    pred = model.predict(sample)[0]

    action = generate_action(pred, sample['lag_1'].values[0], category)

    insight = generate_ai_insight(sample, pred, category)

    return {
        "prediction": float(pred),
        "action": action,
        "insight": insight,
        "category": category
    }
```
